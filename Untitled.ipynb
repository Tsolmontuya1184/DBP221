```python
def usgiin_too(str):
    up_cnt=0
    lo_cnt=0
    for i in str:
        if i.isupper():
            up_cnt+=1
        else:
            lo_cnt+=1
    print(f"Tom usgiin too: {up_cnt}")
    print(f"Jijig usgiin too: {lo_cnt}")
usgiin_too('SeMiNAR')
```

```python
def Urjver(my_list):
    r = 1
    for a in my_list:
        r = r * a
    return r
list = [1, 5, 8, 9]
print(Urjver(list))
```

```python
def factorial(num):
    if num == 1:
        return num
    else:
        return num + factorial(num - 1)
x=int(input("Enter x : "))
y=factorial(x)
print("{}! = {}". format(x,y))
```

```python
def reverse(str):
    s = ""
    for x in str:
        s= x + s
    return s

mystr = "SEMINAR"
print(reverse(mystr))
```

```python
a = [1, 2, 3, 4, 5, 5, 5, 7, 3]
dup_items = set ()
uniq_items = []
for x in a:
    if x not in dup_items:
        uniq_items.append(x)
        dup_items.add(x)
print(uniq_items)
```

```python
def maxhoyr (a,b):
    if a>b:
        return a
    return b
def maxgurav(a,b,c):
    return maxhoyr (a, maxhoyr(b,c))
print(maxgurav(2, 6, 11))
```
